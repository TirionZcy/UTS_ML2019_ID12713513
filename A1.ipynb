1.Introduction
The full name of this paper is “Eigenfaces vs. Fisherfaces: Recognition Using Class Specific Linear Projection” by the authors: Peter N. Belhumeur, Joao~ P. Hespanha, and David J. Kriegman in 1997.

Peter et al’s report deal with a face recognition system that reduces the error rate. They find that the current face recognition system has some problem which is images will deviate from the linear subspace due to the face is not a true Lambertian surface and produce a self-shadow. Therefore, they invented a "Fisherface" method and produced well classes, even under high-intensity illumination. Moreover, a large number of experimental results show that the error rate of their proposed "Fisherface" method is much lower than the Harvard and Yale face database test.

2.Content
2.1 Fundamental knowledge and Challenge
The research is about a new face recognition method which is not sensitive to large changes in light and facial expressions. The light contains not only the light intensity but also the direction and quantity of the light source. The main problems they need to attempted to solve is that some areas of the face will differ between different images. Although the image of the Lambertian surface obtained from the fixed viewpoint is located in the 3D linear subspace of the high dimensional image space under varying illumination, areas of shadows, guesses, and facial expressions often deviate significantly from linear subspace. Therefore, it is an unreliable result for face recognition.
2.2 Four existing face recognition classification methods
In their article, they introduced four existing face recognition classification methods and compare the current popular methods. 

The first method is Correlation. This method identifies the image in the test set by marking the closest point in the test set. If all images are normalized to zero and unit variance, then this process is equivalent to selecting the image associated with the test image. Moreover, in the normalization process, the results are independent of the factors such as the intensity of the light source. The disadvantage of this method is that the study set images and test set images need to be acquired under the same illumination. Therefore, this method cannot get the perfect answer in the case of changes in illumination. The second disadvantage is that the image of the test face must be associated with each image in the study, so the correlation is computationally expensive.

The second method is Eigenfaces. This is a dimensionality reduction method, also known as the Karhunen-Loeve method. It minimizes linear projection by selecting a dimension, maximizing the dispersion of all projection samples and improving accuracy. Its specific expression is as follows:

One disadvantage of this approach is that the interstage scattering due to intra-class scattering used for classification. It means if the lighting changes, most of the features from one image to the next will change.

The third method is Linear Subspaces. The linear subspace algorithm achieves error-free classification under any lighting conditions without noise or shadow, and the surface obeys the Lambertian reflectance model. For each face, three or more images taken in different illumination directions are used to construct the three-dimensional basis of the linear subspace. The disadvantage of this method is that some areas in the face image have variability that they do not have. Linear subspace models need to take into account enough facial images, so you need to know which areas are good for recognition and which are not. Moreover, in order to identify the test image, the distance of each person's linear subspace must be measured. Finally, from a storage perspective, the linear subspace algorithm must preserve three images for each person. It means it will occupy a lot of memory.

The last method is Fisherfaces. The Linear Discriminant Analysis (LDA) theory based on Fisherface and the PCA used in the feature face are similar. They are all methods of mapping the original data to the low-dimensional space. LDA and PCA are all from the data. It is different from LBP to extract local texture features.

2.3 Experimental
In the article, the author mainly compares the above four methods from three aspects which are variation in lighting, variation in facial expression, eye wear, as well as lighting and glasses recognition. The space in the direction of the light source is arranged in a spherical shape and then sampled in increments of 15 degrees.

2.3.1 Variation in lighting
The first experiment was to verify that under variable illumination, if the face recognition algorithm takes advantage of the images of the Lambertian surface, they would work better. The specific experimental method is to determine that the face remains stable and is illuminated by a main light source from each of the selected pictures in the database. The conclusion reached is:
(1)All algorithms work flawlessly when the light is close to the front. However, When the light is off-axis, there is a significant performance difference between correiation and Eigenface. 
(2)When the number of feature faces is equal to the size of the training set, the feature face method is equivalent. Since the performance increases with the dimension of the feature space, the performance of the feature face method should not be better than the correlation.
(3)In the Eigenface method, removing the first three principal components produces better performance under variable lighting conditions.
(4)Although the linear subspaces method has a low error rate, it needs to store more than three times the information and three times the time.
(5)The Fisherface method is more accurate than the Eigenfaces method and has a shorter calculation time.

2.4 Result
(1)If an image similar to the image in the training set is rendered in the test set, all methods perform well. 
(2)The Fisherface method seems to be the best way to extrapolate and interpolate lighting changes.
(3)Removing the three largest principal components does improve the performance of the feature surface method in the case of illumination changes, and does not achieve such a low error rate.
(4)Under certain conditions, as the principal component increases in the feature surface method, its performance is close to correlation.

2.5 Analysis and summary
Peter et al. (1997) state that even with this extensive experiment, there are some interesting questions, such as how the Fisher Faces method extends to large databases. In addition, current face detection methods may reduce the accuracy rate under intense lighting conditions, so new detection methods are needed to support the algorithm. Finally, when shadows occupy most of the face, the performance of all proposed recognition methods will decrease, and techniques for modeling or masking shadow areas may be needed.
